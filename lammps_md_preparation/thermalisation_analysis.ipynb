# Thermalisation analysis

Parse a `log.lammps` file, compute temperature statistics, and visualise the temperature evolution.

This notebook is generic: only the input file path and plot title should need editing.


```python
from pathlib import Path
import sys

cwd = Path.cwd().resolve()
if (cwd / "python" / "lammps_md_preparation" / "__init__.py").exists():
    NOTEBOOK_DIR = cwd
    package_root = cwd / "python"
elif (cwd / "lammps_md_preparation" / "python" / "lammps_md_preparation" / "__init__.py").exists():
    NOTEBOOK_DIR = cwd / "lammps_md_preparation"
    package_root = NOTEBOOK_DIR / "python"
else:
    raise RuntimeError("Run this notebook from the repository root or from lammps_md_preparation/.")

if str(package_root) not in sys.path:
    sys.path.insert(0, str(package_root))

from lammps_md_preparation.utils import plot_temperature_series, read_lammps_temperature_log
```

```python
# Configuration
LOG_FILE = NOTEBOOK_DIR / "input" / "lammps" / "50K_log.lammps"
PLOT_TITLE = "Temperature evolution"
```

```python
timesteps, temperatures = read_lammps_temperature_log(LOG_FILE)
fig, ax, stats = plot_temperature_series(timesteps, temperatures, title=PLOT_TITLE)

print("Temperature statistics:")
for key, value in stats.items():
    if key == "count":
        print(f"- {key}: {int(value)}")
    else:
        print(f"- {key}: {value:.6f}")
```

```python
p25 = stats["p25"]
p75 = stats["p75"]
selected_timesteps = [
    timestep for timestep, temp in zip(timesteps, temperatures)
    if p25 <= temp <= p75
]
print(f"Timesteps inside the interquartile range: {len(selected_timesteps)}")
selected_timesteps[:20]
```
